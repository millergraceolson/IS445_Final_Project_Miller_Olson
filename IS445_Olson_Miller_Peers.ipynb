Depicted below is a line graph showing the performance of top players based on their total yards gained (combining rushing, receiving, and passing yards) across different seasons.

```python
import pandas as pd 
import matplotlib.pyplot as plt
data = pd.read_csv("/Users/millerolson/Desktop/yearly_data_updated_08_23.csv")


data['total_yards'] = data['rushing_yards'] + data['receiving_yards'] + data['passing_yards']
player_total_yards = data.groupby(['name', 'season'])['total_yards'].sum().reset_index()
top_players = player_total_yards.groupby('name')['total_yards'].sum().nlargest(5).index
top_players_data = player_total_yards[player_total_yards['name'].isin(top_players)]
pivot_data = top_players_data.pivot(index='season', columns='name', values='total_yards')
plt.figure(figsize=(12, 6))
pivot_data.plot(marker='o', cmap='tab10', ax=plt.gca())
plt.title('Top Players by Total Yards Gained Over Seasons')
plt.xlabel('Season')
plt.ylabel('Total Yards Gained')
plt.legend(title='Player', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This graph displays the total yards gained by the top players across multiple football seasons. The selection of only five players allows for a focused analysis on the most impactful contributors to offensive production.Each line represents a different player, and the graph illustrates how their total yards gained vary over time. By examining the graph, we can observe trends in player performance, including periods of peak productivity and potential fluctuations in form across different seasons. Additionally, we can identify any notable consistency or variability in the contributions of top players to their teams' offensive output over the years. Overall, this visualization allows us to discern patterns in player performance and gain insights into their impact on the game across various seasons.

Below, I create a pie chart that we can use to show the distribution of total touchdowns among teams for the 2022 season. The percentages in the middle of the pie chart represent the proportion of total touchdowns attributed to each team relative to the total number of touchdowns scored across all teams in the dataset. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame name is 'your_dataframe'

# Filter data for the specific season
season_data = data[data['season'] == 2022]

# Group data by team and calculate total touchdowns
team_total_tds = season_data.groupby('team')['total_tds'].sum()

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(team_total_tds, labels=team_total_tds.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Total Touchdowns by Team (2022 Season)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

The pie chart illustrates the distribution of total touchdowns among different football teams for the 2022 season. Each slice of the pie represents a team, with the size of the slice corresponding to the proportion of touchdowns scored by that team relative to the total touchdowns scored across all teams. The percentages displayed on each slice indicate the contribution of each team to the overall total touchdowns. This visualization offers a quick overview of how touchdown-scoring responsibilities are distributed across teams in the specified season.

```python
season_data = data[data['season'] == 2022]
team_stats = season_data.groupby('team')[['passing_tds', 'rushing_tds']].sum()
team_stats.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Total Passing and Rushing Touchdowns by Team (2022 Season)')
plt.xlabel('Team')
plt.ylabel('Touchdowns')
plt.xticks(rotation=45)
plt.legend(title='Touchdown Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

This stacked bar chart visualizes the total number of passing touchdowns and rushing touchdowns scored by each team during the 2022 season. Each bar represents a team, and the segments within each bar show the contribution of passing touchdowns and rushing touchdowns to the total touchdowns scored by that team. By comparing the lengths of the segments within each bar and between different bars, we can observe the relative distribution of passing and rushing touchdowns across teams, providing insights into the offensive strategies and strengths of each team during the season.
