**<h2>NFL Statistics: General Public</h2>**

This dataset contains comprehensive statistics for various football players across seasons from 2012 to 2022. The data includes metrics such as completions, attempts, passing yards, rushing yards, touchdowns, interceptions, sacks, receiving yards, and many more. These statistics provide detailed insights into the performance of individual players in both passing and rushing aspects of the game, as well as their involvement in receiving. Here is a quick look at the data:

```python
import pandas as pd 
import matplotlib.pyplot as plt
data = pd.read_csv("/Users/millerolson/Desktop/yearly_data_updated_08_23.csv")
data
```

**<h4>Why are NFL statistics so valuable?</h4>**

**<h4>Sports Betting: </h4>** Sports bettors use player and team statistics to inform their betting decisions. Metrics like passing yards, touchdowns, and interceptions can help bettors assess the strengths and weaknesses of teams and predict game outcomes more accurately. Additionally, analyzing historical performance data can reveal trends and patterns that bettors can exploit to identify value bets and maximize their chances of winning.

**<h4>Fantasy Football: </h4>** Fantasy football is a virtual competition where participants act as team owners, drafting real-life NFL players onto their fantasy teams and competing based on the statistical performance of those players in actual NFL games. Fantasy football enthusiasts often rely on player statistics to make informed decisions when drafting players and setting their lineups. Metrics such as touchdowns, receiving yards, and rushing yards are particularly crucial for fantasy football managers, as they directly contribute to the scoring of fantasy points. By analyzing player performance data, fantasy managers can identify top-performing players, predict future performance, and strategize accordingly to build winning fantasy teams.

**<h4>Real Life NFL Draft: </h4>** Football teams and analysts utilize player statistics to evaluate player performance and potential. Metrics such as completion percentage, yards per attempt, and touchdown-to-interception ratio are essential factors considered during player evaluations and draft selections. By analyzing historical data, teams can assess player consistency, skill level, and suitability for their team's needs.

Explain graph below 

```python
data_2021 = data[data['season'] == 2021]
data_2022 = data[data['season'] == 2022]

touchdowns_2021 = data_2021.groupby('team')['total_tds'].sum()
touchdowns_2022 = data_2022.groupby('team')['total_tds'].sum()

teams = touchdowns_2021.index
x = range(len(teams))
bar_width = 0.3

plt.figure(figsize=(100, 8))  # Adjusted figure width

fig, ax = plt.subplots()
bars1 = ax.bar(x, touchdowns_2021, bar_width, label='2021', color='skyblue')
bars2 = ax.bar([i + bar_width for i in x], touchdowns_2022, bar_width, label='2022', color='salmon')

ax.set_xlabel('Team')
ax.set_ylabel('Total Touchdowns')
ax.set_title('Comparison of Total Touchdowns by Team: 2021 vs 2022')
ax.set_xticks([i + bar_width / 2 for i in x])
ax.set_xticklabels(teams, rotation=45, ha='right', fontsize=8)  # Adjusted rotation and font size
ax.legend()

plt.show()
```

The graph compares the total touchdowns scored by each NFL team in the seasons 2021 and 2022. Each bar represents a team, and the height of the bar indicates the total number of touchdowns scored by that team in the respective season. This visualization allows for a quick comparison of team performance in terms of scoring touchdowns between the two seasons.

Below, I created a bar chart showing the average passing yards per game for each team in the 2022 season:

```python

data_2022 = data[data['season'] == 2022]

avg_passing_yards_per_game = data_2022.groupby('team')['passing_yards'].mean()

avg_passing_yards_per_game = avg_passing_yards_per_game.sort_values()

plt.figure(figsize=(10, 6))
avg_passing_yards_per_game.plot(kind='bar', color='green')
plt.title('Average Passing Yards per Game by Team (2022 Season)')
plt.xlabel('Team')
plt.ylabel('Average Passing Yards per Game')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The graph is showing the average passing yards per game by team for the 2022 season in the NFL. Each bar represents a different team, and the height of the bar indicates the average number of passing yards gained per game by that team throughout the season. Teams with higher bars have higher average passing yards per game, while those with lower bars have lower averages. This visualization helps compare the passing performance of different teams in the league during the specified season.

Another visualization is the following pie chart, which shows the distribution of total touchdowns scored by position in the 2022 season:

```python
data_2022 = data[data['season'] == 2022]
total_touchdowns_by_position = data_2022.groupby('position')['total_tds'].sum()
plt.figure(figsize=(8, 8))
total_touchdowns_by_position.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Distribution of Total Touchdowns by Position (2022 Season)')
plt.ylabel('')  # Remove the y-axis label
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.legend(total_touchdowns_by_position.index, loc='best')  # Add a legend with position labels
plt.tight_layout()
plt.show()
```

The pie chart represents the distribution of total touchdowns scored by different player positions (quarterbacks, running backs, wide receivers, and tight ends) in the NFL. Each segment of the pie corresponds to a specific position, and the size of each segment indicates the proportion of total touchdowns attributed to that position. This visualization helps illustrate the contribution of different player positions to the overall touchdown scoring in the league.
